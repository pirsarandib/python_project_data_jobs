```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
```

```python
df = pd.read_csv("data_jobs.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
df.drop(["salary_hour_avg", "job_type_skills", "salary_rate"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df["job_title_short"].value_counts().sort_values().plot(kind="barh", title = "Number of Jobs by Job Title",ylabel="", color='#66b3ff')
plt.legend(["Count"])
```

```python
job_countries = df.groupby(["job_country"]).size().sort_values(ascending=False)
```

```python
job_countries.head(10)
```

```python
median_salary = df.groupby("job_title_short")["salary_year_avg"].median().sort_values(ascending = True)
median_salary.plot(kind="barh", title = "Median Salary by Job Title", ylabel="", xlabel="Salary $",color='#66b3ff')
```

```python
remote_jobs = df[df["job_location"]=="Anywhere"]
non_remote_jobs = df[df["job_location"]!="Anywhere"]
```

```python
grouped_remote_jobs = remote_jobs.groupby('job_title_short').size()
grouped_non_remote_jobs = non_remote_jobs.groupby('job_title_short').size()
```

```python
fix, ax = plt.subplots(1,2, figsize=(12,6))
grouped_remote_jobs.sort_values(ascending=False).head(10).plot(kind="bar",title="Remote Rules",ax=ax[0],xlabel="", color='#66b3ff')
ax[0].set_xticklabels(ax[0].get_xticklabels(), ha='right')
ax[0].tick_params(axis='x', rotation=45)
ax[0].legend(["Count"])

grouped_non_remote_jobs.sort_values(ascending=False).head(10).plot(kind="bar",title="On-site Rules",ax=ax[1],xlabel="", color='#66b3ff')
ax[1].set_xticklabels(ax[1].get_xticklabels(), ha='right')
ax[1].tick_params(axis='x', rotation=45)
ax[1].legend(["Count"])

plt.suptitle("Top 10 Job Titles: Remote vs. On-site Roles")
plt.tight_layout()
plt.show()
```

```python
grouped_remote_jobs_company = remote_jobs.groupby('company_name').size()
grouped_non_remote_jobs_company = non_remote_jobs.groupby('company_name').size()
```

```python
fix, ax = plt.subplots(1,2, figsize=(12,6))
grouped_remote_jobs_company.sort_values(ascending=False).head(10).plot(kind="bar",title="Remote Rules",ax=ax[0],xlabel="", color='#66b3ff')
ax[0].set_xticklabels(ax[0].get_xticklabels(), ha='right')
ax[0].tick_params(axis='x', rotation=45)
ax[0].legend(["Count"])

grouped_non_remote_jobs_company.sort_values(ascending=False).head(10).plot(kind="bar",title="On-site Rules",ax=ax[1],xlabel="", color='#66b3ff')
ax[1].set_xticklabels(ax[1].get_xticklabels(), ha='right')
ax[1].tick_params(axis='x', rotation=45)
ax[1].legend(["Count"])


plt.suptitle("Top 10 Companies: Number of Remote vs. On-site Roles")
plt.tight_layout()
plt.show()
```

```python
degree_mention = df.groupby(["job_title_short","job_no_degree_mention"]).size().unstack(fill_value=0)
```

```python

degree_mention['False_Percentage'] = degree_mention[False] / degree_mention.sum(axis=1) * 100
degree_mention_sorted = degree_mention.sort_values(by='False_Percentage', ascending=False)
degree_mention_sorted
```

```python

degree_mention['False_Percentage'] = degree_mention[False] / degree_mention.sum(axis=1) * 100
degree_mention_sorted = degree_mention.sort_values(by='False_Percentage', ascending=True)
degree_mention_sorted = degree_mention_sorted.drop(columns='False_Percentage')
colors = ['#19ef99','#66b3ff']
degree_mention_sorted.plot(kind="barh", stacked=True, ylabel="", xlabel="Count", color = colors)
plt.title("Percentage of Jobs Mentioning Degree Requirements by Job Title")
```

### for furter analysis I'm only analysing Data Analyst in the US

```python
df_DA = df[(df["job_title_short"] == "Data Analyst") & (df["job_country"] == "United States")]
```

```python
df_DA.head()
```

```python
df_DA.info()
```

```python
fix, ax = plt.subplots(2,1,figsize=(10,10))

sns.kdeplot(data=df_DA, x='salary_year_avg',fill=True, color='#66b3ff', ax=ax[0])
ax[0].set_xlim(0,250000)
ax[0].get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc:f"${int(x/1000)}k"))
ax[0].set_xlabel("")


sns.boxplot(data=df_DA, x ="salary_year_avg",color='#66b3ff' ,ax=ax[1])
ax[1].set_xlim(0,250000)
ax[1].get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc:f"${int(x/1000)}k"))
ax[1].set_xlabel("Salary $")
plt.suptitle('Salary Distribution For Data Analytist')
plt.tight_layout()

plt.show()
```

```python
ls = ["job_schedule_type","job_work_from_home","job_health_insurance"]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 6))
colors = ['#19ef99','#66b3ff']
for index, j in enumerate(ls):
  df_DA[j].value_counts().head(2).plot(kind='pie',ax=ax[index],title=j ,ylabel="", autopct="%.1f%%",startangle=90, colors= colors)

plt.suptitle("Distribution of Job Schedule Type, Work-from-Home Status, and Health Insurance Benefits", fontsize=16)
```
